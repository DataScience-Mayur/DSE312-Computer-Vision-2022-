```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv.imread('/content/cat.jpg')
cv2_imshow(img)

# Print image type
print(img.dtype)

# Convert to grayscale
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)

# Print image size
print(img.shape)

# Rotation
def rotate(img, angle, rotPoint=None):
    (height,width) = img.shape[:2]

    if rotPoint is None:
        rotPoint = (width//2,height//2)

    rotMat = cv.getRotationMatrix2D(rotPoint,angle,1.0)
    dimensions = (width,height)

    return cv.warpAffine(img, rotMat, dimensions)

rotated = rotate(img,-30)
cv2_imshow(rotated)

# Translation
def translate(img, x, y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1],img.shape[0])
    return cv.warpAffine(img,transMat,dimensions)

translated = translate(img,10,20)
cv2_imshow(translated)

# Scaling
def rescaleFrame(frame, scale=0.5):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)

    dimensions = (width, height)

    return cv.resize(frame, dimensions, interpolation=cv.INTER_AREA)

resized_image = rescaleFrame(img)
cv2_imshow(resized_image)

print(resized_image.shape)


# # Rotation image by 30 degree
# h,w = gray.shape
# centre = (h//2,w//2)

# rotation_matrix = cv2.getRotationMatrix2D(centre,30,1.0)
# final_rotated = cv2.warpAffine(gray,rotation_matrix,(w,h))

# # Translate image by [10,20]
# T = np.float32([[1, 0, 10], [0, 1, 20]])
# final_translation = cv2.warpAffine(gray, T, (w,h))

# #Scale image by 1/2 size
# final_scaling = cv2.resize(gray, None, fx = 0.5, fy = 0.5)

# cv2_imshow(final_rotated)
# cv2_imshow(final_translation)
# cv2_imshow(final_scaling)

cv.waitKey(0)
```

```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
import PIL
from PIL import Image, ImageChops 
import matplotlib.pyplot as plt


img = cv2.imread('/content/cat.jpg')
cv2_imshow(img)

smallest = np.amin(img, axis=(0, 1))
largest = np.amax(img, axis=(0, 1))
print(f'Maximum intensity value in original image is {largest}')
print(f'Minimum intensity value in original image is {smallest}')

# Scaling
def rescaleFrame(frame, scale=0.5):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimensions = (width, height)
    return cv2.resize(frame, dimensions, interpolation = cv2.INTER_AREA)

resized_image = rescaleFrame(img)
cv2_imshow(resized_image)

resized_smallest = np.amin(resized_image, axis=(0, 1))
resized_largest = np.amax(resized_image, axis=(0, 1))
print(f'Maximum intensity value in scaled image is {resized_largest}')
print(f'Minimum intensity value in scaled image is {resized_smallest}')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

matrix = Image.fromarray(gray)
print(matrix)

np.mean(matrix)

np.sum(matrix)
type(matrix)

plt.imshow(np.asarray(matrix) + 10, cmap = "gray")

#cv.waitKey(0)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import PIL
from PIL import Image, ImageChops
from scipy.spatial import distance
from google.colab.patches import cv2_imshow

img = cv.imread('/content/cat.jpg')
cv2_imshow(img)

colors = ('b','g','r')

# compute and plot the image histograms
for i, color in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.hist(hist, color=color)

plt.title('Image Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.show()

# Cropping
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv2_imshow(gray)

matrix = Image.fromarray(gray)
I = np.asarray(matrix)

# print(I[0:24])

im = PIL.Image.fromarray(np.uint8(I))
cv2_imshow(I[0:5,0:5])
print(I[0:5,0:5])

cv.waitKey(0)

# Manhattan distance
a = [51, 13]
b = [5, 3]
print(f"Manhattan distance is {distance.cityblock(a,b)}")

# Euclidean distance
print(f"Euclidan distance is {distance.euclidean(a,b)}")

# Chessboard distance
print(f"Chessboard distance is {distance.chebyshev(a,b)}")
```
